```python

import sys
sys.path.append("../src/") 

from graph_eda_features import (
    graph_statistics,
    plot_degree_distribution,
    compute_node_features,
    save_node_features
)

import networkx as nx
import pandas as pd
```

```python
# Step 2: Load preprocessed Facebook graph
fb_file = "../data/processed/facebook_cleaned.edgelist"
G_fb = nx.read_edgelist(fb_file, nodetype=int)

print("Facebook graph loaded:")
print(f"Nodes: {G_fb.number_of_nodes()}, Edges: {G_fb.number_of_edges()}")
```

```python
# Step 3: Facebook Graph EDA
print("Facebook Graph Statistics:")
fb_stats = graph_statistics(G_fb)
print(fb_stats)

# Plot degree distribution
plot_degree_distribution(G_fb)

```

```python
# Step 4: Facebook Node Feature Engineering
fb_features = compute_node_features(G_fb)
print("Sample Facebook node features:")
fb_features.head()

# Save to CSV
save_node_features(fb_features, "../data/processed/facebook_node_features.csv")
```
